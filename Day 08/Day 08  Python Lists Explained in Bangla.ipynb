```python
# Nested list indexing (Positive indexing or negative indexing)

l = [10, 50, 50, 50, 50, 50, ["jak", 80, 90, [1, 2, 3], 800]]
print(l[-1][-2][-2])
```

```python
# List and For Loop Use Case

l = [10, 50, 50, 50, 50, 50, ["jak", 80, 90, [1, 2, 3], 800]]
for i in l:
    print(i)
```

```python
# List filtering

l = [10, 5.5, 5+5j, True, "ridoy", 32, 54, 65.45, 5+4j, False, "inception"]
int_l = []
float_l = []
bool_l = []
str_l = []
complex_l = []

for i in l:
    if type(i) == int:
        int_l.append(i)
    elif type(i) == float:
        float_l.append(i)
    elif type(i) == bool:
        bool_l.append(i)
    elif type(i) == str:
        str_l.append(i)
    else:
        complex_l.append(i)

print(int_l)
print(float_l)
print(bool_l)
print(str_l)
print(complex_l)
```
